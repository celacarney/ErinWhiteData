```r
#Loading in important packages and dataframe
url = 'https://github.com/celacarney/ErinWhiteData/raw/main/bodyfat.csv'
BodyFat = read.csv(url)
library(dplyr)
#library(rstatix)
#install.packages("ggpubr")
```

```r
#Paired t-test for HAWT v HBW
HAW <- BodyFat$Fat_HAWTLCall
HBW <- BodyFat$Fat_HBWTLCall
# Compute t-test
HAW.HBW.Test <- t.test(HAW, HBW, paired = TRUE)
HAW.HBW.Test
```

```r
#Paired t-test for HAWT v RV
HAW <- BodyFat$Fat_HAWTLCall
RV <- BodyFat$Fat_HBWRVall
# Compute t-test
HAW.RV.Test <- t.test(RV, HAW, paired = TRUE)
HAW.RV.Test
```

```r
#Paired t-test for RV v HBW
RV <- BodyFat$Fat_HBWRVall
HBW <- BodyFat$Fat_HBWTLCall
# Compute t-test
HBW.RV.Test <- t.test(RV, HBW, paired = TRUE)
HBW.RV.Test
```

```r
#plot matrix
pairs(BodyFat[, 6:8], xlim=c(0,50), ylim=c(0,50))
```

```r
#individual plots if needed
plot(BodyFat$Fat_HAWTLCall ~ BodyFat$Fat_HBWRVall, xlim=c(0,50), ylim=c(0,50))
plot(BodyFat$Fat_HBWTLCall ~ BodyFat$Fat_HBWRVall, xlim=c(0,50), ylim=c(0,50))
plot(BodyFat$Fat_HBWTLCall ~ BodyFat$Fat_HAWTLCall, xlim=c(0,50), ylim=c(0,50))
```

```r
#means
meanHAW <- mean(BodyFat$Fat_HAWTLCall)
meanHBW <- mean(BodyFat$Fat_HBWTLCall)
meanRV <- mean(BodyFat$Fat_HBWRVall, na.rm = TRUE)
meanRV
meanHBW
meanHAW
```
